```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
cancer['target_names']
```

```python
cancer['feature_names']
```

```python
X = cancer['data'][:, 0]
X = X.reshape((X.shape[0], 1))
```

```python
y = cancer['target']
y = 1 - y.reshape((y.shape[0], 1))
```

```python
plt.scatter(X, y)
```

Linear Regression?

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X, y)
```

```python
plt.scatter(X, y);
plt.plot(X, model.predict(X));
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X, y.ravel())
```

```python
plt.scatter(X, y);
plt.scatter(X, model.predict_proba(X)[:, 1], marker='.')
```

```python
plt.scatter(np.reshape(X, -1), np.reshape(y, -1), c=np.reshape(model.predict(X), -1), cmap=plt.cm.rainbow, marker='.');
plt.scatter(X, model.predict_proba(X)[:, 1], marker='.');
plt.axhline(.5);
plt.axvline(-model.intercept_ / model.coef_);
```
